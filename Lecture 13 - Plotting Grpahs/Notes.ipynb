# Matplotlib

The `matplotlib.pyplot.plot()` function in Python is used to plot lines and/or markers. 
### Basic Syntax

```python
matplotlib.pyplot.plot(*args, **kwargs)
```

Note: Positional Arguments (*args)

Note: Keyword Arguments (**kwargs)

### Positional Arguments (`*args`)

`plt.plot()` can take multiple forms depending on how you pass the data. Common forms include:

- `plot(y)` – x is automatically set to range(len(y))
- `plot(x, y)` – x and y are data points
- `plot(x, y, format_string)` – format string controls color, marker, and line


### Keyword Arguments (`**kwargs`)

These control the **style and appearance** of the plot. Here's a categorized list of the most commonly used keyword arguments:

#### Line Properties

- `color` or `c` – line color (`'blue'`, `'g'`, `'#FFDD44'`, etc.)
- `linestyle` or `ls` – (`'-'`, `'--'`, `'-.'`, `':'`, `'None'`)
- `linewidth` or `lw` – line width in points
- `marker` – marker style (`'o'`, `'s'`, `'D'`, `'^'`, `'x'`, etc.)
- `markersize` or `ms` – marker size in points
- `markeredgecolor` or `mec` – marker edge color
- `markerfacecolor` or `mfc` – marker face color
- `markeredgewidth` or `mew` – marker edge width

#### Alpha and Antialiasing

- `alpha` – transparency (0 to 1)
- `antialiased` or `aa` – boolean for smooth lines

#### Dash Properties

- `dash_capstyle`
- `dash_joinstyle`
- `dashes` – custom dash pattern

#### **Markers with Customization**

- `marker` – style (as above)
- `markevery` – used to mark every nth data point
- `fillstyle` – `'full'`, `'left'`, `'right'`, `'bottom'`, `'top'`, `'none'`

#### **Labeling and Z-order**

- `label` – used for legend
- `zorder` – layer ordering

#### **Data and Transform**

- `data` – pass a dict or a DataFrame
- `transform` – coordinate transform

````python
# Basic Example
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
````


The `matplotlib.pyplot.scatter()` function is used to create scatter plots in Python.

## Basic Syntax

```python
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, 
                          vmin=None, vmax=None, alpha=None, linewidths=None, 
                          edgecolors=None, plotnonfinite=False, *, data=None, **kwargs)
```

## Parameters Explained

### Required Arguments

- `x` – *array-like*: x-coordinates of the data points
- `y` – *array-like*: y-coordinates of the data points

### Appearance Customization

#### Marker Style and Size

- `marker` – marker symbol (`'o'`, `'s'`, `'^'`, `'*'`, `'x'`, `'D'`, etc.)
- `s` – size of markers (can be scalar or array-like)
- `edgecolors` – color of marker edges
- `linewidths` – width of marker edges

#### Color

- `c` – color(s) of markers (scalar, array-like, or color names)
- `cmap` – colormap (e.g., `'viridis'`, `'plasma'`, `'coolwarm'`, etc.)
- `norm` – Normalize object for scaling `c`
- `vmin`, `vmax` – value limits for colormap normalization
- `alpha` – transparency (0 to 1)


### Data Handling

- `data` – dict or DataFrame to reference `x`, `y`, `s`, `c` by string names
- `plotnonfinite` – whether to plot points with NaN or inf values


### Keyword Arguments

- Other `kwargs` are passed to `matplotlib.collections.PathCollection` (advanced usage like picking, offsets, etc.)


## Examples

### 1. Basic Scatter Plot

```python
plt.scatter([1, 2, 3], [4, 5, 6])
```


### 2. Colored and Sized Points

```python
plt.scatter(x, y, s=100, c='red', marker='^')
```


### 3. Colormap Based on Value

```python
colors = [10, 20, 30, 40]
sizes = [100, 200, 300, 400]
plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7)
plt.colorbar()
```


### 4. Edge Colors and Transparency

```python
plt.scatter(x, y, s=150, edgecolors='black', linewidths=2, alpha=0.5)
```

The `matplotlib.pyplot.pie()` function is used to create pie charts in Python.


## Basic Syntax

```python
matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None,
                      pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0,
                      radius=1, counterclock=True, wedgeprops=None, textprops=None,
                      center=(0, 0), frame=False, rotatelabels=False, normalize=None,
                      *, data=None)
```

## Parameters Explained

### Required

- `x` – *array-like*: The wedge sizes (numerical values for each slice)


### Labeling and Annotation

- `labels` – *list of strings*: Labels for each wedge
- `autopct` – Format string or function to label the wedges with their numeric value (e.g., `'%1.1f%%'`)
- `pctdistance` – Distance of the `autopct` text from the center (default `0.6`)
- `labeldistance` – Distance of the labels from the center (default `1.1`)
- `rotatelabels` – Rotate each label to match slice angle (default `False`)


### Style and Appearance

- `colors` – *list of colors*: Slice colors (e.g., `['red', 'blue']`)
- `explode` – *array-like*: Fraction of the radius with which to offset each wedge
- `shadow` – *bool*: Draw a shadow beneath the pie (default `False`)
- `startangle` – *float*: Start angle in degrees (default `0`)
- `radius` – Radius of the pie chart (default `1`)
- `counterclock` – Plot slices counterclockwise (default `True`)
- `frame` – Whether to draw a frame around the pie chart


### Properties

- `wedgeprops` – Dict of properties for the wedges (e.g., `{'linewidth': 2, 'edgecolor': 'black'}`)
- `textprops` – Dict of properties for the text (e.g., `{'fontsize': 14, 'color': 'red'}`)


### Advanced

- `center` – Tuple of floats (x, y): Center of the pie chart (default `(0, 0)`)
- `normalize` – Normalize x to sum to 1. If `False`, uses raw values (default `None`)
- `data` – DataFrame or dict, if using string keys for `x`, `labels`, etc.


## Returns

```python
patches, texts, autotexts = plt.pie(...)
```

````python
# Basic Pie chart
import matplotlib.pyplot as plt
plt.pie([1,2,3,4])
plt.axis('equal')
plt.show()
````

````python
# With Labels and Explode
import matplotlib.pyplot as plt
sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']
explode = [0.1, 0, 0, 0]

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=180)
plt.axis('equal') # To make a perfect circle
plt.show()
````

````python
# With Custom Colors and Text Properties
import matplotlib.pyplot as plt
plt.pie([25, 25, 25, 25],
        labels=['Q1', 'Q2', 'Q3', 'Q4'],
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],
        textprops={'fontsize': 14, 'color': 'black'})
plt.axis("equal")
plt.show()
````

The `matplotlib.pyplot.hist()` function is used to create **histograms** in Python using Matplotlib.

## Basic Syntax
```python
matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None,
                       cumulative=False, bottom=None, histtype='bar', align='mid',
                       orientation='vertical', rwidth=None, log=False, color=None,
                       label=None, stacked=False, *, data=None, **kwargs)
```

## Parameters Explained

### Required

- `x` – *array-like*: Input data to be histogrammed.

### Bin Configuration

- `bins` – int, sequence, or string:

  - Number of bins (e.g., `10`)
  - Bin edges (e.g., `[0, 1, 2, 3]`)
  - Or method like `'auto'`, `'fd'`, `'sturges'`, `'doane'`, `'scott'`, `'rice'`, `'sqrt'`
- `range` – Tuple of (min, max) to limit the range of bins.

### Calculation Behavior

- `density` – *bool*: Normalize histogram to form a probability density (default `False`).
- `weights` – *array-like*: Weights for each data point.
- `cumulative` – *bool or -1*: Compute cumulative histogram (`True`, `False`, `-1` for reverse).

### Display Customization

- `bottom` – *array-like or scalar*: Starting position of bars on the y-axis.
- `histtype` – Type of histogram:

  - `'bar'` (default): Traditional bar-type
  - `'barstacked'`: Stacked bars
  - `'step'`: Line plot without filling
  - `'stepfilled'`: Line plot with filling
- `align` – `'left'`, `'mid'` (default), `'right'`: Alignment of bar edges.
- `orientation` – `'vertical'` (default) or `'horizontal'`.

### Style

- `rwidth` – Relative width of bars (0 to 1)
- `log` – *bool*: Plot on a log scale (default `False`)
- `color` – Color or list of colors
- `label` – String or list of labels
- `stacked` – *bool*: If `True`, multiple datasets are stacked

### Other

- `data` – Dataset (e.g., dictionary or DataFrame)
- `**kwargs` – Passed to `PatchCollection` (e.g., `edgecolor`, `linewidth`, etc.)

## Returns

```python
n, bins, patches = plt.hist(...)
```

- `n`: Values of histogram bins
- `bins`: Edges of the bins
- `patches`: List of bar objects (for styling)

````python
# Basic Histogram
import matplotlib.pyplot as plt
plt.hist([1,2,3,4,5,6])
plt.show()
````

````python
# With Custom Bins and Color
import matplotlib.pyplot as plt
plt.hist([1,2,3,4,5,6],bins=20,color="skyblue",edgecolor='black')
plt.show()
````

````python
# Normalized (Density) and Cumulative Histogram
import matplotlib.pyplot as plt
plt.hist([1,2,3,4,5],bins=21,density=True,cumulative=True,color="yellow")
plt.show()
````

````python
# Horizontal Histogram
import matplotlib.pyplot as plt
plt.hist([1,2,3,4,5,6],orientation="horizontal", color="orange")
plt.show()
````

````python
# Multiple Datasets (Stacked)
import numpy as np
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)
plt.hist([x1, x2], bins=30, stacked=True, color=['blue', 'green'])
plt.show()
````
