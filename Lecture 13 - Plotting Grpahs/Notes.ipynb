# Matplotlib

The `matplotlib.pyplot.plot()` function in Python is used to plot lines and/or markers. 
### Basic Syntax

```python
matplotlib.pyplot.plot(*args, **kwargs)
```

Note: Positional Arguments (*args)

Note: Keyword Arguments (**kwargs)

### Positional Arguments (`*args`)

`plt.plot()` can take multiple forms depending on how you pass the data. Common forms include:

- `plot(y)` – x is automatically set to range(len(y))
- `plot(x, y)` – x and y are data points
- `plot(x, y, format_string)` – format string controls color, marker, and line


### Keyword Arguments (`**kwargs`)

These control the **style and appearance** of the plot. Here's a categorized list of the most commonly used keyword arguments:

#### Line Properties

- `color` or `c` – line color (`'blue'`, `'g'`, `'#FFDD44'`, etc.)
- `linestyle` or `ls` – (`'-'`, `'--'`, `'-.'`, `':'`, `'None'`)
- `linewidth` or `lw` – line width in points
- `marker` – marker style (`'o'`, `'s'`, `'D'`, `'^'`, `'x'`, etc.)
- `markersize` or `ms` – marker size in points
- `markeredgecolor` or `mec` – marker edge color
- `markerfacecolor` or `mfc` – marker face color
- `markeredgewidth` or `mew` – marker edge width

#### Alpha and Antialiasing

- `alpha` – transparency (0 to 1)
- `antialiased` or `aa` – boolean for smooth lines

#### Dash Properties

- `dash_capstyle`
- `dash_joinstyle`
- `dashes` – custom dash pattern

#### **Markers with Customization**

- `marker` – style (as above)
- `markevery` – used to mark every nth data point
- `fillstyle` – `'full'`, `'left'`, `'right'`, `'bottom'`, `'top'`, `'none'`

#### **Labeling and Z-order**

- `label` – used for legend
- `zorder` – layer ordering

#### **Data and Transform**

- `data` – pass a dict or a DataFrame
- `transform` – coordinate transform

````python
# Basic Example
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
````

````python
# Line with Custom Color and Style
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, color='green', linestyle='--', linewidth=2)
plt.show()
````

````python
# Line with Markers
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker="^")
plt.show()
````

````python
# Line with Custom Markers and Marker Size
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker="s", markersize=10, markerfacecolor='red',markeredgecolor='black')
plt.show()
````

````python
# Line with Format String ('ro--' = red, circle, dashed)
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, "ro--")
plt.show()
````

````python
# Multiple Lines with Labels
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
x2 = [1, 2, 3, 4]
y2 = [40, 30, 20, 10]
plt.plot(x,y,label="Line 1")
plt.plot(x2,y2,label="Line 2")

plt.legend()
plt.show()
````

````python
# Custom Dash Pattern
import matplotlib.pyplot as plt

line, = plt.plot(x, y, color='purple')
line.set_dashes([5, 2, 10, 5])  # 5 on, 2 off, 10 on, 5 off
plt.show()
````

````python
# Line with Transparency
import matplotlib.pyplot as plt
plt.plot(x, y, color='red', alpha=0.5)
plt.show()
````

````python
# Plotting with Data from Dictionary
import matplotlib.pyplot as plt
data = {'x': [1, 2, 3], 'y': [4, 5, 6]}
plt.plot(data["x"], data["y"], label='Dict Plot',linestyle='-')
plt.legend()
plt.show()
````

````python
# Plotting with Data with grid
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,3,2,1])
plt.grid()
plt.show()
````


The `matplotlib.pyplot.scatter()` function is used to create scatter plots in Python.

## Basic Syntax

```python
matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, 
                          vmin=None, vmax=None, alpha=None, linewidths=None, 
                          edgecolors=None, plotnonfinite=False, *, data=None, **kwargs)
```

## Parameters Explained

### Required Arguments

- `x` – *array-like*: x-coordinates of the data points
- `y` – *array-like*: y-coordinates of the data points

### Appearance Customization

#### Marker Style and Size

- `marker` – marker symbol (`'o'`, `'s'`, `'^'`, `'*'`, `'x'`, `'D'`, etc.)
- `s` – size of markers (can be scalar or array-like)
- `edgecolors` – color of marker edges
- `linewidths` – width of marker edges

#### Color

- `c` – color(s) of markers (scalar, array-like, or color names)
- `cmap` – colormap (e.g., `'viridis'`, `'plasma'`, `'coolwarm'`, etc.)
- `norm` – Normalize object for scaling `c`
- `vmin`, `vmax` – value limits for colormap normalization
- `alpha` – transparency (0 to 1)


### Data Handling

- `data` – dict or DataFrame to reference `x`, `y`, `s`, `c` by string names
- `plotnonfinite` – whether to plot points with NaN or inf values


### Keyword Arguments

- Other `kwargs` are passed to `matplotlib.collections.PathCollection` (advanced usage like picking, offsets, etc.)


## Examples

### 1. Basic Scatter Plot

```python
plt.scatter([1, 2, 3], [4, 5, 6])
```


### 2. Colored and Sized Points

```python
plt.scatter(x, y, s=100, c='red', marker='^')
```


### 3. Colormap Based on Value

```python
colors = [10, 20, 30, 40]
sizes = [100, 200, 300, 400]
plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7)
plt.colorbar()
```


### 4. Edge Colors and Transparency

```python
plt.scatter(x, y, s=150, edgecolors='black', linewidths=2, alpha=0.5)
```

````python
# example of creating x^3 graph
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,5,0.1)
y1 = x**2
y2 = x**3
y3 = x**4


plt.title("matplotlib example")

plt.plot(x,y1, label="x**2")
plt.plot(x,y2, label="x**3")
plt.plot(x,y3, label="x**4")
plt.legend()

plt.axis([0,10,0,1000])
plt.grid()
plt.text(2,100,"2,100")
plt.show()
````
