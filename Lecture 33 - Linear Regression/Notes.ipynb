# Linear Regression

Predict student marks based on hours studied.

Step 1: Dataset

| Hours Studied | Marks Scored |
| ------------- | ------------ |
| 1             | 50           |
| 2             | 55           |
| 3             | 65           |
| 4             | 70           |
| 5             | 75           |
| 6             | 85           |


Step 2: Code for Linear Regression using Scikit-learn


```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Step 2: Prepare data
data = {
    'Hours': [1, 2, 3, 4, 5, 6],
    'Marks': [50, 55, 65, 70, 75, 85]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Step 3: Split into X (input) and y (output)
X = df[['Hours']]   # Features must be 2D Reason FEATURES can be many
y = df['Marks']     # Labels can be 1D Reason OUTPUT should be one
```

```python
# Step 4: Train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 5: Predict marks for a new value (e.g., 7 hours)
predicted = model.predict([[7]])
print(f"Predicted marks for 7 hours of study: {predicted[0]:.2f}")
```

```python
# Step 6: Plot the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

# Coefficient of Determination or R² (R-squared)

What is the Coefficient of Determination (R²)?
- The Coefficient of Determination (R²) is a metric used to evaluate how well a regression model fits the data.
- It measures the proportion of the variance in the dependent variable (y) that is predictable from the independent variable(s) (X).


Simple Definition:
- R² tells you how much of the variation in the output (y) is explained by your model.
    - R² = 1 → Perfect fit (100% of the variation is explained by the model)
    - R² = 0 → The model explains none of the variation
    - R² < 0 → The model is worse than just predicting the average

Note: R² is used to check the performance of a regression model.

Formula

$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$


$$
y_i = actual-observed-value
$$

$$
\bar{y} = mean-of-actual-observed-values
$$

$$
\hat{y_i} = predicted-value-from-the-model
$$

$$
{n} = number-of-data-points
$$

# Cost Function

What is a Cost Function?
- A cost function is a mathematical function that measures how wrong the model’s predictions are.
- It quantifies the error between the model’s predicted values and the actual (true) values.

Why is it important?
- The goal of training a model is to minimize the cost function.
- By reducing this cost, the model learns to make better predictions.



Common Cost Functions (Regression):

1. Mean Squared Error (MSE) – most common for linear regression
2. Mean Absolute Error (MAE) – alternative for regression
3. Binary Cross-Entropy (Log Loss) - Logistic regression
