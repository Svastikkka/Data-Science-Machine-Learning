Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

#### Problem Statement :
Find value of killed column only where `country == ‘United States’`?

Print `0` in place of missing values.

Print `count` of Killed as integer value.

#### Output Format :
```
Killed1
Killed2
Killed3
Killed4
. . .
. . .
. . .
```

````python
import pandas as pd

# Load the dataset from the CSV file
data = pd.read_csv("../Datasets/terrorismData.csv")

# Print the list of column names to identify the correct column name for 'country'
# print(data.columns)

# Assuming the column name is 'Country', replace it with the correct column name if needed
killed_in_usa = data.loc[data['Country'] == 'United States', 'Killed']

# Replace missing values with 0
killed_in_usa.fillna(0, inplace=True)

# Print the killed values as integer
for killed in killed_in_usa:
    print(int(killed))

````
